```python
import numpy as np
import scipy as sp
```

```python
from scipy.stats import randint , norm
# randint for uniform probability - mass function 
# norm for density function
```

```python
# probability mass function for event from 1 - 6 [will see what's the prob of '1']

a = 1
b = 6

prob = randint.pmf(1 , a ,b+1)
print(prob)
```

```python
# probability density function (not the probability)
a = 10
density = norm.pdf(a)
```

```python
# density function for random values
x = np.linspace(-5,5)
y = sp.stats.norm.pdf(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# x = values , y = density 
plt.plot(x,y)
```

```python
# comulative distribution - real probability
# Note that: every time we run without mentioning any other parameter, scipy put mean = 0 and std =1 
X = norm.cdf(x)
plt.plot(x,X)
```
