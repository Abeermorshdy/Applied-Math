## Central limit theory

- We will apply CLT on exponintial distribution. We will declare that the big sample > 30 will produce normal sampling distribution. 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
lambda_ = 3
d = np.random.exponential(1/lambda_,size=1000)
sns.histplot(d,kde=True)
```

```python
means = []
lambda_ = 3
for i in range(1000):
    d = np.random.exponential(1/lambda_,size=1000)
    mu = np.mean(d)
    means.append(mu)

print(f"population mean :{1/lambda_}")
print(f"sample distribution mean :{np.mean(means)}")
sns.histplot(means,kde=True)
```

- the mean of population is the same as the mean of sample distribution

- The minimum sample size is 30. If you choose less than 30, THE distribution of sampling size will not be normal  

- If you apply this on any distribution its sample bigger than 30, it will generate sample ditribution with normal distribution.

- For variance, as the size of sample will increase, the variance will decrease

```python
means = []
lambda_ = 3
for i in range(1000):
    d = np.random.exponential(1/lambda_,size=1000)
    mu = np.mean(d)
    means.append(mu)

print(f"variance: {np.var(means)}")
print(f"population mean :{1/lambda_}")
print(f"sample distribution mean :{np.mean(means)}")
sns.histplot(means,kde=True)
```
