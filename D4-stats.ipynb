#### 3. Poisson distribution
- In this distribution, you care about the rate of occurance of something `lambda`, and the mean `mu` will equal to `lambda`

```python
from scipy.stats import poisson 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
lambda_ = 2
poisson.pmf(k = 2 , mu = lambda_ )
```

```python
lambda_ = 2
poisson.pmf(k = 1,mu = lambda_ )
```

```python
lambda_ = 2
poisson.pmf(k = 3,mu = lambda_ )
```

```python
x  = np.arange(0,6)
p = poisson.pmf(k = x,mu = 2 )
sns.barplot(x = x , y = p)
```

## Continoues Distributions 

#### 1. Uniform continuos distribution

```python
from scipy.stats import uniform
a = 10
b = 30
mu , segma = uniform.stats(loc = a, scale = b-a, moments = 'mv')
print(mu,segma)
print(uniform.cdf(x = 20 ,loc = a, scale = b-a) - uniform.cdf(x = 15 , loc = a, scale = b-a))
```

```python
x = np.arange(10,30)
pdf = uniform.pdf(x , loc = a ,scale = b-a)   # Area under the curve
cdf = uniform.cdf(x , loc = a , scale = b-a)  # Integrated area
plt.plot(x,pdf)
plt.show()
plt.plot(x,cdf)
```

```python
a = 10
b = 30
data = np.random.uniform(a, b, size = 10)
print('mean= ', data.mean() , 'variance= ', data.var())
sns.histplot(x= data)
```

```python
a = 10
b = 30
data = np.random.uniform(a, b, size = 1000)
print('mean= ', data.mean() , 'variance= ', data.var())
sns.histplot(x= data)
```

```python
a = 10
b = 30
data = np.random.uniform(a, b, size = 100000)
print('mean= ', data.mean() , 'variance= ', data.var())
sns.histplot(x= data)
```

#### 2. Exponential distribution

![Alt text](image.png)

```python
from scipy.stats import expon
lambda_1 = 1/50000
lambda_2 = 1/20000
mean_1 = expon.stats(loc = 0, scale = 1/lambda_1 , moments = 'm')
mean_2 = expon.stats(loc = 0, scale = 1/lambda_2 , moments = 'm')
print(mean_1, mean_2)
```

```python
x = np.linspace(0,200000)
p_1 = expon.pdf(x,loc = 0, scale = 1/lambda_1)
p_2 = expon.pdf(x,loc = 0, scale = 1/lambda_2)
plt.plot(x,p_1)
plt.plot(x,p_2)
plt.show()
```

```python
p_1 = expon.cdf(x,loc = 0, scale = 1/lambda_1)
p_2 = expon.cdf(x,loc = 0, scale = 1/lambda_2)
plt.plot(x,p_1)
plt.plot(x,p_2)
plt.show()
```

```python
data_1 = np.random.exponential(scale=1/lambda_1 , size = 100)
data_2 = np.random.exponential(scale=1/lambda_2 , size = 100)
sns.histplot(x = data_1,kde=True)
sns.histplot(x = data_2,kde=True)
```

```python
data_1 = np.random.exponential(scale=1/lambda_1 , size = 10000)
data_2 = np.random.exponential(scale=1/lambda_2 , size = 10000)
sns.histplot(x = data_1,kde=True)
sns.histplot(x = data_2,kde=True)
```

```python
from scipy.stats import norm
norm.pdf(1,loc = 0 , scale = 1)
```

```python
# this is the probability from - infinity to 1 std
norm.cdf(1,loc = 0 , scale = 1)
```

```python
# if we want to know the probability from -1 to 1 std
norm.cdf(1,loc = 0 , scale = 1) - norm.cdf(-1,loc = 0 , scale = 1)
```

```python
# if we want to know the probability from -2 to 2 std
norm.cdf(2 ,loc = 0 , scale = 1) - norm.cdf(-2 ,loc = 0 , scale = 1)
```

```python
# if we have the probability and want to figure out what is std we can use `ppt` Percent point function which 
# count the probablity from - infinity to the probability you count before
norm.ppf(0.8413447460685429 , loc = 0, scale = 1)
```

```python
# if you want to count the standared deviation in which 95% of the data in
norm.ppf(0.5 + (0.95/2), loc = 0, scale = 1)
```

- if we want to know how far the point x from the mean we can calculate the normalization for each point [(x-mean)/std]  
- The standared deviation values which we calculated from the previous operation is also called `Z-Scores` or `Z-distributions`
