## Discrete distributions

#### 3. Poisson distribution
- In this distribution, you care about the rate of occurance of something `lambda`, and the mean `mu` will equal to `lambda`

```python
from scipy.stats import poisson 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
lambda_ = 2
poisson.pmf(k = 2 , mu = lambda_ )
```

```python
lambda_ = 2
poisson.pmf(k = 1,mu = lambda_ )
```

```python
lambda_ = 2
poisson.pmf(k = 3,mu = lambda_ )
```

```python
x  = np.arange(0,6)
p = poisson.pmf(k = x,mu = 2 )
sns.barplot(x = x , y = p)
```

```python
x  = np.arange(0,11)
p = poisson.pmf(k = x,mu = 2 )
sns.barplot(x = x , y = p)
```

```python
x  = np.arange(0,20)
p = poisson.pmf(k = x,mu = 10 )
sns.barplot(x = x , y = p)
```

- Note:
    - when you increase `lambda`, you will find the the shape will be more normal 

Let's take a sample from numpy and see what we will get

```python
data  = np.random.poisson(lam = 2, size = 10)
```

```python
data.mean()
```

- we should get exactly 2, as the mean always equal to lambda. If we increase the size of sample we will get almost 2

## Continoues Distributions 

#### Uniform continuos distribution

```python
from scipy.stats import uniform
a = 10
b = 30
mu , segma = uniform.stats(loc = a, scale = b-a, moments = 'mv')
print(mu,segma)
print(uniform.cdf(x = 20 ,loc = a, scale = b-a) - uniform.cdf(x = 15 , loc = a, scale = b-a))
```

```python
x = np.arange(10,30)
pdf = uniform.pdf(x , loc = a ,scale = b-a)   # Area under the curve
cdf = uniform.cdf(x , loc = a , scale = b-a)  # Integrated area
plt.plot(x,pdf)
plt.show()
plt.plot(x,cdf)
```

```python
a = 10
b = 30
data = np.random.uniform(a, b, size = 10)
print('mean= ', data.mean() , 'variance= ', data.var())
sns.histplot(x= data)
```

```python
a = 10
b = 30
data = np.random.uniform(a, b, size = 1000)
print('mean= ', data.mean() , 'variance= ', data.var())
sns.histplot(x= data)
```

```python
a = 10
b = 30
data = np.random.uniform(a, b, size = 100000)
print('mean= ', data.mean() , 'variance= ', data.var())
sns.histplot(x= data)
```
